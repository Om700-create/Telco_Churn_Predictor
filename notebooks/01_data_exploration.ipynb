### Multicollinearity Check

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Calculate VIF for numerical features
vif_data = pd.DataFrame()
vif_data['Feature'] = numeric_df.columns
vif_data['VIF'] = [variance_inflation_factor(numeric_df.values, i) for i in range(numeric_df.shape[1])]

# Display VIF results
vif_data.sort_values(by='VIF', ascending=False, inplace=True)
print(vif_data)
```

Observations:

1. VIF values help to identify features that are highly correlated with each other.
2. A VIF above 5-10 indicates multicollinearity that may need addressing in your model.


### Correlation of Categorical Features with Target

```python
# Analyze how categorical features correlate with the target variable (Churn) using pivot tables
# Create a DataFrame to hold churn rates for each categorical feature
churn_rate = {}

# Loop through each categorical feature and calculate churn rates
for feature in ['gender', 'Partner', 'Dependents', 'InternetService']:
    churn_counts = df.groupby(feature)['Churn'].value_counts(normalize=True).unstack().fillna(0)
    churn_rate[feature] = churn_counts

# Convert the churn_rate dictionary into a DataFrame for better display
churn_rate_df = pd.concat(churn_rate)
churn_rate_df.columns = ['Not Churned', 'Churned']
print(churn_rate_df)
```

Observations:
1. The code above constructs the churn rates for each categorical feature correctly.
2. The pd.concat() function is used to combine the results for better readability.
3. This will allow you to see the churn rates for each group within the categorical features, helping you identify potential patterns related to customer churn.

### Identify Possible Outliers

```python
# Using IQR to identify outliers in Total Charges
Q1 = df['TotalCharges'].quantile(0.25)
Q3 = df['TotalCharges'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['TotalCharges'] < lower_bound) | (df['TotalCharges'] > upper_bound)]
print(f"\nNumber of Outliers in Total Charges: {outliers.shape[0]}")
```

1. Identifying outliers can inform data cleaning decisions and ensure that the analysis reflects the customer base accurately.
2. This step is essential for improving model performance and accuracy.


###  Churn Rate by Contract Type and Tenure

```python

# Create a pivot table for churn rate based on Contract and Tenure
churn_rate_contract_tenure = pd.pivot_table(df, values='Churn', 
                                            index=pd.cut(df['tenure'], bins=[0, 12, 24, 48, 60, 72]), 
                                            columns='Contract', 
                                            aggfunc=lambda x: x.value_counts(normalize=True).get(1, 0))

# Fill NaN with 0 for visualization
churn_rate_contract_tenure = churn_rate_contract_tenure.fillna(0)

# Visualize the pivot table using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(churn_rate_contract_tenure, annot=True, cmap='YlOrRd', fmt=".2f")
plt.title('Churn Rate by Contract Type and Tenure')
plt.xlabel('Contract Type')
plt.ylabel('Tenure Range (in months)')
plt.show()
```

1. This heatmap reveals how tenure length and contract type affect churn.
2. We can expect that customers with longer tenures and yearly contracts may be less likely to churn.
